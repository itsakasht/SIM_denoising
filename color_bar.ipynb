```python
import numpy as np
import optics as op
import matplotlib.pyplot as plt
import skimage as ski
import os
```

```python
original = ski.color.rgb2gray(ski.io.imread('source_images/Synthetic512.png'))

img = ski.io.imread("output_images/poisson_noise/10percent_noise/tv_synthetic_grated_256_sim.tif")
img1 = ski.io.imread("output_images/poisson_noise/10percent_noise/tv_synthetic_grated_256_sim_display.tif")
image = ski.util.img_as_uint(img) # uint16
# image = img1

fig, ax = plt.subplots(1, 3, figsize=(15, 3.5))

ax[0].imshow(img[:, :, 0], cmap='gray')
ax[0].set_title('TV filter & 10% poisson noise (float)')
ax[0].axis('off')
plt.colorbar(ax[0].imshow(img[:, :, 0], cmap='gray'))

ax[1].imshow((image[:, :, 0]), cmap='gray')
ax[1].set_title('TV filter & 10% poisson noise (uint16)')
ax[1].axis('off')
plt.colorbar(ax[1].imshow((image[:, :, 0]), cmap='gray'))

ax[2].imshow(original, cmap='gray')
ax[2].set_title('Original Image')
ax[2].axis('off')
plt.colorbar(ax[2].imshow(original, cmap='gray'))

plt.show()

print(np.min(image[:, :, 0]), np.max(image[:, :, 0]), np.mean(image[:, :, 0]))

imgnorm = op.normalize(image[:, :, 0])
print(np.min(imgnorm), np.max(imgnorm), np.mean(imgnorm))
print(np.min(original), np.max(original), np.mean(original))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 3.5))
ax[0].hist(img[:, :, 0].flatten(), bins=64)
ax[0].set_title('TV filter & 10% poisson noise (float)')
ax[0].set_ylim(0, 180000)

ax[1].hist(imgnorm.flatten(), bins=64)
ax[1].set_title('TV filter & 10% poisson noise (uint16)')
ax[1].set_ylim(0, 180000)

ax[2].hist(original.flatten(), bins=64)
ax[2].set_title('Original Image')
ax[2].set_ylim(0, 180000)

plt.show()
```

```python
fig4data = ski.io.imread("source_images/Figure4Data.tif")
fig4data_sim = ski.io.imread("source_images/Figure4Data_sim.tif")[:, :, 0]

# fig4data0 = np.mean(fig4data, axis=0)
fig4data0 = fig4data[0]
print(np.min(fig4data0), np.max(fig4data0), np.mean(fig4data0))
print(np.min(fig4data_sim), np.max(fig4data_sim), np.mean(fig4data_sim))

plt.imshow(fig4data0, cmap='gray')
plt.title('Figure 4 Data')
plt.axis('off')
plt.colorbar(plt.imshow(fig4data0, cmap='gray'))
plt.show()

# fig4data_sim = op.normalize(fig4data_sim)
plt.imshow(fig4data_sim, cmap='gray')
plt.title('Figure 4 Data SIM')
plt.axis('off')
plt.colorbar(plt.imshow(fig4data_sim, cmap='gray'))
plt.show()
```
